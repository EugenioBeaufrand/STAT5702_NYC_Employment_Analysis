```python
import pandas as pd
import numpy as np
import os
```

```python
TotalNY = pd.read_csv("C:/Users/eabea/Dropbox/Documents/STAT W5702/Project/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/education/Total_NY.csv")
```

```python
TotalNY.style
```

```python
years = [i for i in range(2010,2020)]


PropEducatedBySalary = pd.DataFrame(columns = ['year','Salary','PropEdu'])

for y in years:
    
    N_Edu = np.sum(TotalNY[TotalNY['year'] == y]['NumEducated'])
    N_Bach = np.sum(TotalNY[(TotalNY.Education_Level == "Bechelor's Degree or Higher") & (TotalNY.year == y)]['NumEducated'])
    PropEducated = N_Bach / N_Edu
    PropFemale = np.sum(TotalNY[(TotalNY.year == y) & (TotalNY.Gender == 'Female')]['NumEducated']) / np.sum(TotalNY[TotalNY['year'] == y]['NumEducated'])
    PropMale = 1 - PropFemale
    
    WeightedAvgSalary = PropMale * np.mean(TotalNY[(TotalNY.year == y) & (TotalNY.Gender == 'Male')]['Salary_CountyGender']) + PropFemale * np.mean(TotalNY[(TotalNY.year == y) & (TotalNY.Gender == 'Female')]['Salary_CountyGender'])
    
    PropEducatedBySalary = PropEducatedBySalary.append(pd.Series([y,WeightedAvgSalary,PropEducated],index=['year', 'Salary', 'PropEdu']),ignore_index=True)
```

```python
PropEducatedBySalary.to_csv(os.path.join('C:/Users/eabea/Dropbox/Documents/STAT W5702/Project/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/Education/','PropEducatedBySalary.csv'),index = False)
```

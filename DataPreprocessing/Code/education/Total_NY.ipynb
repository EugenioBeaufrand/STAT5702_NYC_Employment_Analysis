```python
import pandas as pd
import numpy as np
import os
```

```python
NumEmpData = pd.read_csv(r"/Users/tracy/Documents/GitHub/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/numemp_sum_races_combined.csv")
CountySalaryEduData = pd.read_csv(r"/Users/tracy/Documents/GitHub/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/education/salary_edu.csv")
EduData = pd.read_csv(r"/Users/tracy/Documents/GitHub/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/education/Education.csv")

CountySalaryEduData = CountySalaryEduData.drop(columns = ["Unnamed: 0"])
CountySalaryEduData['year'] = CountySalaryEduData['year'].str.replace('year','')
NumEmpData = NumEmpData.drop(columns = ["Unnamed: 0"])

BronxSalaryEduDataM = CountySalaryEduData.iloc[:10,:].reset_index(drop = True)
BronxSalaryEduDataF= CountySalaryEduData.iloc[10:20,:].reset_index(drop = True) 
QueensSalaryEduDataM = CountySalaryEduData.iloc[20:30,:].reset_index(drop = True) 
QueensSalaryEduDataF = CountySalaryEduData.iloc[30:40,:].reset_index(drop = True) 
KingsSalaryEduDataM = CountySalaryEduData.iloc[40:50,:].reset_index(drop = True) 
KingsSalaryEduDataF = CountySalaryEduData.iloc[50:60,:].reset_index(drop = True) 
New_YorkSalaryEduDataM = CountySalaryEduData.iloc[60:70,:].reset_index(drop = True) 
New_YorkSalaryEduDataF = CountySalaryEduData.iloc[70:80,:].reset_index(drop = True) 
RichmondSalaryEduDataM = CountySalaryEduData.iloc[80:90,:].reset_index(drop = True) 
RichmondSalaryEduDataF = CountySalaryEduData.iloc[90:100,:].reset_index(drop = True) 
CountySalaryEduData['year'] = CountySalaryEduData['year'].str.replace('year','')

EduData
```

```python
EduData = EduData.iloc[:,[1,6,7,9,10,12,13,15,16,18,19,20]]

EduData['Male'] = EduData['Bronx County, Male']+EduData['Kings County, Male']+EduData['New York County, Male']+EduData['Queens County, Male']+EduData['Richmond County, Male']
EduData['Female'] = EduData['Bronx County, Female']+EduData['Kings County, Female']+EduData['New York County, Female']+EduData['Queens County, Female']+EduData['Richmond County, Female']

EduData = EduData.iloc[:,[0,11,12,13]]

EduData.to_csv("/Users/tracy/Documents/GitHub/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/education/NY_Edu.csv")
EduData
```

```python
BronxEmpM = NumEmpData.iloc[[1,4,7,10,13,16,19,22,25,28],[2,7]]
BronxEmpM.reset_index(inplace = True, drop = True)
BronxEmpM = pd.concat([BronxEmpM, BronxSalaryEduDataM],axis=1)
BronxEmpM = BronxEmpM.rename(columns={'Bronx': 'numemp'})
BronxEmpM['Salary_BronxGender'] = BronxEmpM['Salary_CountyGender']*BronxEmpM['numemp']
BronxEmpM = BronxEmpM.iloc[:,[1,2,4,6]]

BronxEmpM
```

```python
QueensEmpF = NumEmpData.iloc[[2,5,8,11,14,17,20,23,26,29],[5,7]]
QueensEmpF.reset_index(inplace = True, drop = True)
QueensEmpF = pd.concat([QueensEmpF, QueensSalaryEduDataF],axis=1)
QueensEmpF = QueensEmpF.rename(columns={'Queens': 'numemp'})
QueensEmpF['Salary_QueensGender'] = QueensEmpF['Salary_CountyGender']*QueensEmpF['numemp']
QueensEmpF = QueensEmpF.iloc[:,[1,2,4,6]]


QueensEmpF
```

```python
QueensEmpM = NumEmpData.iloc[[1,4,7,10,13,16,19,22,25,28],[5,7]]
QueensEmpM.reset_index(inplace = True, drop = True)
QueensEmpM = pd.concat([QueensEmpM, QueensSalaryEduDataM],axis=1)
QueensEmpM = QueensEmpM.rename(columns={'Queens': 'numemp'})
QueensEmpM['Salary_QueensGender'] = QueensEmpM['Salary_CountyGender']*QueensEmpM['numemp']
QueensEmpM = QueensEmpM.iloc[:,[1,2,4,6]]


QueensEmpM
```

```python
KingsEmpM = NumEmpData.iloc[[1,4,7,10,13,16,19,22,25,28],[3,7]]
KingsEmpM.reset_index(inplace = True, drop = True)
KingsEmpM = pd.concat([KingsEmpM, KingsSalaryEduDataM],axis=1)
KingsEmpM = KingsEmpM.rename(columns={'Kings': 'numemp'})
KingsEmpM['Salary_KingsGender'] = KingsEmpM['Salary_CountyGender']*KingsEmpM['numemp']
KingsEmpM = KingsEmpM.iloc[:,[1,2,4,6]]


KingsEmpM
```

```python
KingsEmpF = NumEmpData.iloc[[2,5,8,11,14,17,20,23,26,29],[3,7]]
KingsEmpF.reset_index(inplace = True, drop = True)
KingsEmpF = pd.concat([KingsEmpF, KingsSalaryEduDataF],axis=1)
KingsEmpF = KingsEmpF.rename(columns={'Kings': 'numemp'})
KingsEmpF['Salary_KingsGender'] = KingsEmpF['Salary_CountyGender']*KingsEmpF['numemp']
KingsEmpF = KingsEmpF.iloc[:,[1,2,4,6]]


KingsEmpF
```

```python
New_YorkEmpF = NumEmpData.iloc[[2,5,8,11,14,17,20,23,26,29],[4,7]]
New_YorkEmpF.reset_index(inplace = True, drop = True)
New_YorkEmpF = pd.concat([New_YorkEmpF, New_YorkSalaryEduDataF],axis=1)
New_YorkEmpF = New_YorkEmpF.rename(columns={'New_York': 'numemp'})
New_YorkEmpF['Salary_New_YorkGender'] = New_YorkEmpF['Salary_CountyGender']*New_YorkEmpF['numemp']
New_YorkEmpF = New_YorkEmpF.iloc[:,[1,2,4,6]]


New_YorkEmpF
```

```python
New_YorkEmpM = NumEmpData.iloc[[1,4,7,10,13,16,19,22,25,28],[4,7]]
New_YorkEmpM.reset_index(inplace = True, drop = True)
New_YorkEmpM = pd.concat([New_YorkEmpM, New_YorkSalaryEduDataM],axis=1)
New_YorkEmpM = New_YorkEmpM.rename(columns={'New_York': 'numemp'})
New_YorkEmpM['Salary_New_YorkGender'] = New_YorkEmpM['Salary_CountyGender']*New_YorkEmpM['numemp']
New_YorkEmpM = New_YorkEmpM.iloc[:,[1,2,4,6]]


New_YorkEmpM
```

```python
RichmondEmpF = NumEmpData.iloc[[2,5,8,11,14,17,20,23,26,29],[6,7]]
RichmondEmpF.reset_index(inplace = True, drop = True)
RichmondEmpF = pd.concat([RichmondEmpF, RichmondSalaryEduDataF],axis=1)
RichmondEmpF = RichmondEmpF.rename(columns={'Richmond': 'numemp'})
RichmondEmpF['Salary_RichmondGender'] = RichmondEmpF['Salary_CountyGender']*RichmondEmpF['numemp']
RichmondEmpF = RichmondEmpF.iloc[:,[1,2,4,6]]


RichmondEmpF
```

```python
RichmondEmpM = NumEmpData.iloc[[1,4,7,10,13,16,19,22,25,28],[6,7]]
RichmondEmpM.reset_index(inplace = True, drop = True)
RichmondEmpM = pd.concat([RichmondEmpM, RichmondSalaryEduDataM],axis=1)
RichmondEmpM = RichmondEmpM.rename(columns={'Richmond': 'numemp'})
RichmondEmpM['Salary_RichmondGender'] = RichmondEmpM['Salary_CountyGender']*RichmondEmpM['numemp']
RichmondEmpM = RichmondEmpM.iloc[:,[1,2,4,6]]


RichmondEmpM
```

```python
Male = pd.merge(BronxEmpM,QueensEmpM,on = ["New York City","Gender","year"])
Male = pd.merge(Male,RichmondEmpM,on = ["New York City","Gender","year"])
Male = pd.merge(Male,New_YorkEmpM,on = ["New York City","Gender","year"])
Male = pd.merge(Male,KingsEmpM,on = ["New York City","Gender","year"])

Male['Salary_CountyGender'] = (Male['Salary_BronxGender']+Male['Salary_QueensGender']+Male['Salary_RichmondGender']+Male['Salary_New_YorkGender']+Male['Salary_KingsGender'])/Male['New York City']
Male = Male.iloc[:,[0,1,2,8]]
Male = Male.rename(columns={'New York City': 'NumEmp'})

Male
```

```python
Female = pd.merge(BronxEmpF,QueensEmpF,on = ["New York City","Gender","year"])
Female = pd.merge(Female,RichmondEmpF,on = ["New York City","Gender","year"])
Female = pd.merge(Female,New_YorkEmpF,on = ["New York City","Gender","year"])
Female = pd.merge(Female,KingsEmpF,on = ["New York City","Gender","year"])

Female['Salary_CountyGender'] = (Female['Salary_BronxGender']+Female['Salary_QueensGender']+Female['Salary_RichmondGender']+Female['Salary_New_YorkGender']+Female['Salary_KingsGender'])/Female['New York City']
Female = Female.iloc[:,[0,1,2,8]]
Female = Female.rename(columns={'New York City': 'NumEmp'})

Female
```

```python
edu_salary = pd.concat([Male,Female],axis=0)
edu_salary
```

```python
import numpy as np
import pandas as pd
import os
```

```python
Data = pd.read_csv('C:/Users/eabea/Dropbox/Documents/STAT W5702/Project/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/Occupation_Yearly.csv')
```

```python
Data.style
```

```python
Occupations = []
i = 0

while Data.iloc[i][2] == 2010:
    Occupations.append(Data.iloc[i][1])
    i += 1

Averages = {}

for occupation in Occupations:
    
    Averages[occupation] = float((Data[Data['Occupations']==occupation]['YearlyNumEmp']).mean())
    
Averages = pd.DataFrame.from_dict(Averages,orient = 'index')
Averages.reset_index(inplace = True)

print(Averages)
```

```python
Averages.to_csv(os.path.join('C:/Users/eabea/Dropbox/Documents/STAT W5702/Project/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/','AverageEmpsByOccupation.csv'),index = True)
```

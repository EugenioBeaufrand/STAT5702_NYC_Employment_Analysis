```python
import pandas as pd
import numpy as np
import os
```

```python
Data = pd.read_csv('/Users/tracy/Desktop/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/numemp_combined.csv')
Data = Data.drop(columns = 'Unnamed: 0')

Data
```

```python

numemp_big_category = Data[(Data['Occupations'] == 'Management, business, science, and arts occupations:')|
                           (Data['Occupations'] == 'Service occupations:')|
                           (Data['Occupations'] == 'Sales and office occupations:')|
                           (Data['Occupations'] == 'Natural resources, construction, and maintenance occupations:')|
                           (Data['Occupations'] == 'Production, transportation, and material moving occupations:')]
numemp_big_category['Bronx County, New York'] = Data.groupby(['Occupations','year'])['Bronx County, New York'].transform('sum')
numemp_big_category['Kings County, New York'] = Data.groupby(['Occupations','year'])['Kings County, New York'].transform('sum')
numemp_big_category['New York County, New York'] = Data.groupby(['Occupations','year'])['New York County, New York'].transform('sum')
numemp_big_category['Queens County, New York'] = Data.groupby(['Occupations','year'])['Queens County, New York'].transform('sum')
numemp_big_category['Richmond County, New York'] = Data.groupby(['Occupations','year'])['Richmond County, New York'].transform('sum')

numemp_big_category = numemp_big_category.iloc[:,[0,1,2,3,4,5,6]].drop_duplicates()
                         
numemp_big_category.to_csv('/Users/tracy/Desktop/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/numemp_big_category.csv')
```

```python
import pandas as pd
import numpy as np
import os
```

```python
Data = pd.read_csv('/Users/tracy/Desktop/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/AggregateYearlyData.wide.csv')
Data
```

```python
Data.index.values
Years = [i for i in range(1,10)]

def add_variance(Years_data):
    variance = 0
    for year in Years:
        variance_temp = abs(Years_data[year+1] - Years_data[year])
        if (variance_temp > variance):
            variance = variance_temp
    return variance

Data['range'] = Data.apply(add_variance,axis = 1)
Data['range_pct'] = Data['range']/((Data["2010"] + Data["2011"] + Data["2012"] + Data["2013"] + Data["2014"] + Data["2015"] + Data["2016"] + Data["2017"] + Data["2018"] + Data["2019"])/10)
Data = Data.iloc[:,[0,11,12]]
Data
```

```python
Data.to_csv(r"/Users/tracy/Desktop/STAT5702_NYC_Employment_Analysis/DataPreprocessing/Data/NumEmp/Occupation_Yearly_withRange.csv",index = False)
```
